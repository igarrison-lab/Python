```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# Making the function
def main(any_string):
    # Capitalize the letters
    any_string = any_string.upper()
    # Split the string
    any_string = any_string.split()
    # Separating the words into different variables
    first_letter = any_string[0]
    second_letter = any_string[1]
    third_letter = any_string[2]
    # Extracting the first letters
    first_letter = first_letter[0:1]
    second_letter = second_letter[0:1]
    third_letter = third_letter[0:1]
    # Printing the output
    print(first_letter, second_letter, third_letter, sep = ". ")

# calling the striong
if __name__ == "__main__":
    main("John William Smith")
```

### Problem 8

#### Question Text

Write a program with a function that accepts a string as an argument and returns a copy of the string with the first character of each sentence capitalized. For instance, if the argument is “hello. my name is Joe. what is your name?” the function should return the string "Hello. My name is Joe. What is your name?" The program should let the user enter a string and then pass it to the function. The modified string should be displayed.

#### High-Level Overview

First I had to ask for the string and capitalize the first letter before anything. Then, at first I tried to use find but the index counting was too complicated so I instead split them by periods. I looped through the separated tokens, capitalizing the first letter and then joining them, undoing the split. Overall, would say this one was a bit complicated.


```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# Creating the function
def main():
    # ask user for a string
    user_string = input("Please enter multiple sentences.")
    # capitalize the first letter
    user_string = user_string.capitalize()
    # split them into sentences
    sentences = user_string.split(". ")
    # capitalize every sentence
    for i in range(len(sentences)):
        sentences[i] = sentences[i].capitalize()
    # joining them back together
    user_string = ". ".join(sentences)
    # display results
    print(user_string)


if __name__ == "__main__":
    main()
```

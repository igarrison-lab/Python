#### Question Text

Write a program that asks the user for a number in the range of 1 through 7. The program should display the corresponding day of the week, where 1 = Monday, 2 = Tuesday, 3 = Wednesday, 4 = Thursday, 5 = Friday, 6 = Saturday, and 7 = Sunday. The program should display an error message if the user enters a number that is outside the range of 1 through 7.

#### High-Level Overview

I first tried to figure out what I had to do and how to do it. I first thought to ask the user for a number so I made the input. I then had to make it so it prints the day based off a number so I used the if statements. It was an overall pretty simple concept.
```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# Ask user to pick a number
number = int(input("Please pick a number 1 through 7:   "))

# Find and print the weekday based off number
if number == 1:
    print("Monday")
elif number == 2:
    print("Tuesday")
elif number == 3:
    print("Wednesday")
elif number == 4:
    print("Thursday")
elif number == 5:
    print("Friday")
elif number == 6:
    print("Saturday")
elif number == 7:
    print("Sunday")
else:
    print("error: you have picked a number outside the range")
    
```

### Problem 5

#### Question Text

Scientists measure an object’s mass in kilograms and its weight in newtons. If you know the amount of mass of an object in kilograms, you can calculate its weight in newtons with the following formula:
weight = mass x 9.8
Write a program that asks the user to enter an object’s mass, then calculates its weight. If the object weighs more than 500 newtons, display a message indicating that it is too heavy. If the object weighs less than 100 newtons, display a message indicating that it is too light.

#### High-Level Overview

First, I of course had to get the user's input for the mass for the rest of the problem. After that, I calculated the equation based off the operator formula from last chapter. I then followed the rest of the problem and displayed the message if it was too light or if it was too heavy as asked. Overall, very straightforwar.


```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# Get mass from user
mass = float(input("Please enter an object's mass in kilograms"))

# Calculate the weight
weight = mass * 9.8

# Display message if too heavy
if weight > 500:
    print("That weight is too heavy")

# Display message if too light
if weight < 100:
    print("That weight is too light")

```

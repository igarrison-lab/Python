Assume a file containing a series of integers is named numbers.txt and exists on the computer’s disk. Write a program that displays all of the numbers in the file.

This one is simple, using the past codes to go off, I made a loop of numbers for the text file. I then open it again with read this time and used the tutorial to convert to integers and display.


```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# Making the number file
with open("numbers.txt", "w") as output_file:
    for line in range(1,10):
        output_file.write("1\n")
        

# Making program that converts to integers, then prints
with open("numbers.txt", "r") as input_file:
    for line in input_file:
        number = int(line)
        print(number)
```

### Problem 6

#### Question Text

Assume a file containing a series of integers is named numbers.txt and exists on the computer’s disk. Write a program that calculates the average of all the numbers stored in the file.

#### High-Level Overview

This one is much more difficult. First, this one is similar to no.1 so I basically copied the first part over. I eventually figured out I had to Calculate the total and count first, then find average after the loop. he operations weren't hard at least.


```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# Setting total and count variables before loop
total = 0
count = 0

# Calculating sum and counts
with open("numbers.txt", "r") as input_file:
    for line in input_file:
        number = int(line)
        total += number
        count += 1

# After finding total and count, find average
average = total / count
print(f"The average total is {average}")
```

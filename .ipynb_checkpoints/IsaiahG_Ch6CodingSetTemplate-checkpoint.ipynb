### Problem 1

#### Question Text

Write a program that opens an output file with the filename my_name.txt, writes your name to the file, then closes the file. 

#### High-Level Overview

This is getting very cool. The difficulty was not bad, the chapter overview walked me through it so no mistakes. I first asked for the name, then used the with open syntax to write it.


```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# Asking user for name
user_name = input("What is your name?")

# Opening and writing the file
with open ("my_name.txt", "w") as output_file:
     output_file.write(user_name)
```

### Problem 2

#### Question Text

Write a program that opens the my_name.txt file that was created by the program in problem 1, reads your name from the file, displays the name on the screen, then closes the file.

#### High-Level Overview

This one was a bit more difficult. I didn't see an exact syntax for this setup so I just used a variable to output the read command to. The question also says to print the name on screen so I add the print funtion. Overall, pretty chill and not bad.

```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# Opening, reading, and displaying the file
with open ("my_name.txt", "r") as input_file:
    content = input_file.read()
    print(content)
```

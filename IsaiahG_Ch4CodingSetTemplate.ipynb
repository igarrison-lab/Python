### Problem 1

#### Question Text

A bug collector collects bugs every day for five days. Write a program that keeps a running total of the number of bugs collected during the five days. The loop should ask for the number of bugs collected for each day, and when the loop is finished, the program should display the total number of bugs collected.

#### High-Level Overview

I tried to think a bit outside the box on this one. Since it was 5 days, I made the loop around the days variable and add a day each time. I also had to think of how I were to get the total and I thought of adding another variable called total to dump the number too. The only thing I really messed up on was forgetting to set the total variable to 0 at first. In the end, it worked out.


```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# Setting the days
days = 1

# Setting the total amount
total = 0

# Making the loop for bugs collected
while days < 6:
    bugs = float(input(f"please enter the number of bugs collected on day {days}"))
    total += bugs
    days += 1

# Display the total
print (f"The total amount of bugs collected is {total}")
```

### Problem 4

#### Question Text

The distance a vehicle travels can be calculated as follows:
distance = speed x time
For example, if a train travels 40 miles per hour for three hours, the distance traveled is 120 miles. Write a program that asks the user for the speed of a vehicle (in miles per hour) and the number of hours it has traveled. It should then use a loop to display the distance the vehicle has traveled for each hour of that time period. Here is an example of the desired output: 


#### High-Level Overview

I tried to keep this question simple. I of course asked for the speed and hours first. I then has to figure out how to use the loop to calculate the equation. I wasn't sure if you could use another vairable in the range argument but it worked. As it is, I thought to use the for loop to display distance for each hour, exactly as intended. Not many other problems other than than.


```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# Ask user for speed of the vehicle in mph
speed = float(input("What is the speed of the vehicle in mph?  "))

# Ask user for hours traveled
hours = int(input("How many hours has it traveled?  "))

# Calculate and display using loop
for hours in range(hours + 1):
    distance = speed * hours
    print(f"On hour {hours} your distance is {distance} miles")
```

For example, assume the user enters the following values:
Starting number of organisms: 2
Average daily increase: 30%
Number of days to multiply: 10

#### High-Level Overview

```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# ask user for starting number of organisms
organisms = int(input("Please enter starting number of organisms:  "))

# ask user for average daily population increase
increase = float(input("Please enter the average daily population increase (as a percentage):  "))

# ask for number of days the organisms will be left to multiply
days = int(input("Please enter the amount of days the organisms will be left to multiply:  "))

# calculating and displaying the data
for days in range(1, days + 1):
    print(f"On day {days}, you will have approximately {organisms} organisms")
    percent = organisms * (increase/100)
    organisms += percent
```

### Problem 1

#### Question Text

Design a program that asks the user to enter a store’s sales for each day of the week. The amounts should be stored in a list. Use a loop to calculate the total sales for the week and display the result.

#### High-Level Overview

This one is pretty simple. I also realized that I should be making functions to reduce global variables so I am gonna start doing that. I first asked for the sales for the week. I then made the list and loop to calculate total just like the previous question.


```python
# add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

def main():
    # Ask user for sales during the week
    monday_sales = float(input("Please enter the store's sales for Monday:  "))
    tuesday_sales = float(input("Please enter the store's sales for Tuesday:  "))
    wednesday_sales = float(input("Please enter the store's sales for Wednesday:  "))
    thursday_sales = float(input("Please enter the store's sales for Thursday:  "))
    friday_sales = float(input("Please enter the store's sales for Friday:  "))
    saturday_sales = float(input("Please enter the store's sales for Saturday:  "))
    sunday_sales = float(input("Please enter the store's sales for Sunday:  "))
    # Creating the list
    weekly_sales = [monday_sales, tuesday_sales, wednesday_sales, thursday_sales, friday_sales, saturday_sales, sunday_sales]
    # Setting the value for total
    total = 0
    for i in range(len(weekly_sales)):
        total += weekly_sales[i]
    return total

print(f"The total sales for the week is {main()}")
```

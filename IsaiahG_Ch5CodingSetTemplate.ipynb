### Problem 16

#### Question Text

A prime number is a number that is only evenly divisible by itself and 1. For example, the number 5 is prime because it can only be evenly divided by 1 and 5. The number 6, however, is not prime because it can be divided evenly by 1, 2, 3, and 6.338                            

Write a Boolean function named  which takes an integer as an argument and returns true if the argument is a prime number, or false otherwise. Use the function in a program that prompts the user to enter a number then displays a message indicating whether the number is prime.

#### High-Level Overview

This one was much more rough than the last questions. I took a lot of inspo from professor becuase I ddin't know the equation on how to check for prime numbers well. I made it so you can actually leave the loop tho by finding the prime number and added the sq rt limiter for quicker execution.


```python
###### add your code to this cell
# remember to include comments like this one for most of your code
# remember to "Run cell" once you're done, excecuting your code so the output is displayed below

# importing the square root limitor
import math

# Making the prime number checker
def is_prime(num):
    limit = int(math.sqrt(num))
    for i in range(2, limit + 1):
        is_evenly_divisible = (num % i == 0)
        if is_evenly_divisible:
            return False
    return True


# Test user input
def main():
    while True:
        number = int(input("Please enter a number to find if it's a prime number"))
        prime_result = is_prime(number)
        if prime_result:
            print(f"your number {number} is a prime number")
            break
        else:
            print(f"The number {number} is not a prime number")

# Run input tester
main()
```
